# 4. 对于以下贝叶斯网络：





 - P(A=1)
 - P(B=1|D=1)?
 - P(B=1|A=1)?
 - P(B=1|A=1,D=1)?

解：
 - P(A=1) = W0*B0*A1+W0*B1*A1+W1*B1*A1+W1*B0*A1


```python
W0 = 0.4
W1 = 0.6
B0 = 0.99
B1 = 0.01

result = W0*B0*0.01+W0*B1*0.9+W1*B1*0.95+W1*B0*0.6

print(result)
```

 - P(B=1|D=1) = P(B=1) = 0.01

 - P(B=1|A=1) = B1*A1 / {B1*A1+B0*A1} 



```python
FENZI = W1*B1*0.95+W0*B1*0.9

FENMU = W0*B0*0.01+W1*B0*0.6+FENZI

result = FENZI/FENMU

print(result)
```
