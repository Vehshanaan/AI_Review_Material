# 2. Confusion Matrix如下所示：


- precision高于recall?
- F1?

```python
# Precision: TP/all marked as P

Precision = 90/91

# Recall: TP/所有实际的P

Recall = 90/95

print("Precision: "+str(Precision))
print("Recall: "+str(Recall))

# F1 = 2/(1/pre +1/rec)

F1 = 2*(Precision*Recall)/(Precision+Recall)

print("F1: "+str(F1))
```
